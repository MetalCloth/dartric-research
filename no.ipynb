```python
from sqlalchemy.schema import CreateTable
from sqlalchemy import MetaData

# 1. Load all table definitions from the database into Python
metadata = MetaData()
metadata.reflect(bind=engine)

# 2. Loop through every table and print its "CREATE" statement
# .sorted_tables ensures they are printed in order of dependencies (parents first)
for table in metadata.sorted_tables:
    # The .compile(engine) part ensures it uses specific PostgreSQL syntax
    print(CreateTable(table).compile(engine))
    print("-" * 50)
```
