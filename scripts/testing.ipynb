````python
def get_table_ddl():
    inspector = inspect(engine)
    table_names = inspector.get_table_names(schema='public')
    
    print("---  DATABASE FOUNDATION (DDL) ---")
    
    for table in table_names:
        print(f"\nCREATE TABLE {table} (")
        columns = inspector.get_columns(table, schema='public')
        
        # Loop through columns to mimic the SQL format
        lines = []
        for col in columns:
            col_str = f"  {col['name']} {col['type']}"
            if not col['nullable']:
                col_str += " NOT NULL"
            lines.append(col_str)
            
        # Join them with commas
        print(",\n".join(lines))
        print(");")
        print("-" * 30)

        break
````

````python
from sqlalchemy import create_engine, MetaData
from sqlalchemy.schema import CreateTable
````

````python
def dump_schema():
    try:
        engine = create_engine(DB_URL)
        metadata = MetaData()
        
        # Fetch table definitions
        metadata.reflect(bind=engine)
        print(f" Connected! Found {len(metadata.tables)} tables.")
        
        # Open file to write
        with open("my_schema.sql", "w", encoding="utf-8") as f:
            for table_name in metadata.tables:
                
                # --- FILTERS ---
                # Skip the AI tables (fixes NullType error)
                if table_name in ['embeddings', 'embeddings_documents']:
                    continue
                # Skip the Partition tables (noise)
                # if "payment_p" in table_name:
                #     continue
                # Skip internal Postgres tables
                if table_name.startswith("pg_"):
                    continue

                # Generate DDL
                table = metadata.tables[table_name]
                ddl = CreateTable(table).compile(engine)
                
                # Write to file
                f.write(f"-- Table: {table_name}\n")
                f.write(str(ddl).strip() + ";\n\n")
                print(f"   Saved: {table_name}")

        print("\n DONE! Check the file 'my_schema.sql' in this folder.")

    except Exception as e:
        print(f" Error: {e}")
        print(" Hint: If 'Name not known', change aws-0 to aws-1 in the db_host variable.")
````
